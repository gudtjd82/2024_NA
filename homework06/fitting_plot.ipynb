```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fitted_params(x, y, a):
    xp_fit = a[0][0] * x + a[1][0] * y + a[2][0]
    yp_fit = a[0][1] * x + a[1][1] * y + a[2][1]
    return xp_fit, yp_fit
```

# Data 1

```python
data = np.loadtxt("./fitdata/fitdata1.dat")
x, y, xp, yp = data[:, 0], data[:, 1], data[:, 2], data[:, 3]
```

```python
a = np.array([
    [0.490944, 0.000625],
    [0.001270, 0.491081],
    [-0.187587, 0.578860],
])
```

```python
xp_fit, yp_fit = fitted_params(x, y, a)
```

```python
residuals_x = xp - xp_fit
residuals_y = yp - yp_fit

mean_residual_x = np.mean(residuals_x)
std_residual_x = np.std(residuals_x)

mean_residual_y = np.mean(residuals_y)
std_residual_y = np.std(residuals_y)

print(f"Mean Residual (x'): {mean_residual_x}, Std Dev: {std_residual_x}")
print(f"Mean Residual (y'): {mean_residual_y}, Std Dev: {std_residual_y}")
```

```python
plt.figure(figsize=(12, 6))

# x' 
plt.subplot(1, 2, 1)
plt.plot(xp, label="Actual x'", linestyle='dotted', color='blue')
plt.plot(xp_fit, label="Fitted x'", linestyle='solid', color='red')
plt.xlabel("Data Point Index")
plt.ylabel("x'")
plt.legend()
plt.title("Comparison of x' (Actual vs Fitted)")

# y' 
plt.subplot(1, 2, 2)
plt.plot(yp, label="Actual y'", linestyle='dotted', color='blue')
plt.plot(yp_fit, label="Fitted y'", linestyle='solid', color='red')
plt.xlabel("Data Point Index")
plt.ylabel("y'")
plt.legend()
plt.title("Comparison of y' (Actual vs Fitted)")

plt.tight_layout()
plt.show()
```
